```python
#导入库
import pyecharts
import pandas as pd
import numpy as np
from pyecharts import options as opts
from pyecharts.charts import Bar,Timeline
from pyecharts.faker import Faker
import re
import os
import time
from pyecharts.globals import ThemeType
```

```python
tit='16.csv'
fea_data = pd.read_csv("E:\\desktop\\data\\fea\\[特征]"+tit, encoding='gbk')
name_list=list(fea_data['name'])
name_set=list(set(name_list))
fea_list=list(fea_data.columns)
fea_list.remove('name')


y_dict={}
std_std={}
for fea in fea_list:
    count={1:[],2:[],3:[]}
    fea_std=[]
    for name in name_set:
        data=fea_data[fea_data['name']==name]
        data=list(data[fea])
        fea_std.append(np.std(data))
        for i in range(len(data)):
            count[i+1].append(data[i])
    y_dict[fea]=count
    std_std[fea]=np.std(fea_std)

std_std = dict(sorted(std_std.items(), key=lambda x: x[1],reverse=False))#正序
fea_list=list(std_std.keys())
std_std
```

```python
tl = Timeline(init_opts=opts.InitOpts(theme=ThemeType.WALDEN))#theme=ThemeType.WALDEN)
for fea in fea_list:
    c = (
        Bar(init_opts=opts.InitOpts(theme=ThemeType.WALDEN))
        .add_xaxis(name_set)
        .add_yaxis("第一次",y_dict[fea][1])#, stack="stack1"
        .add_yaxis("第二次", y_dict[fea][2])
        .add_yaxis("第三次", y_dict[fea][3])
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title=fea),
            yaxis_opts=opts.AxisOpts(
                                    min_=0,
                                    max_=1,
                                    )
        )
    )

    tl.add(c, fea)
tl.render_notebook()

```
